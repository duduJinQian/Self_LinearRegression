```python
## 数据加载
ytrue = np.load('ytrue1.npy')
xreal = np.load('xreal.npy')
weightinit = np.load('weightinit.npy')
```

```python
### 划分训练集 及 验证集
import random

def getRandomIndex(n, x):# 索引范围为[0, n), 随机选x个不重复
    index = random.sample(range(n), x)
    return index

lengthy=len(ytrue)

trainindex=np.array(getRandomIndex(1284, 900))
valindex=np.delete([x for x in range(0,lengthy)], trainindex)

xtrian=xreal[trainindex]
ytrian=ytrue[trainindex]

xval=xreal[valindex]
yval=ytrue[valindex]
```

```python
### 模型训练及获取结果
bbb=linear.Linear(xtrian,ytrian,xval,yval,weightinit,learningrate=0.001,nstep=8000)
weightnew,lossend=bbb()
```
